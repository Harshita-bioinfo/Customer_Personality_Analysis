# Data Cleaning and Preprocessing

# 1. Outliers Detection and Handling them

```python
# Creates and displays boxplots for all numeric columns in the marketing_data DataFrame to visualize potential outliers, with customized layout and spacing.

# Identify all numeric columns (excluding object types)
num_columns = marketing_data.select_dtypes(exclude='object').columns

# Set layout for plotting boxplots
plots_per_row = 4
total_rows = (len(num_columns) + plots_per_row - 1) // plots_per_row

# Increase figure size for more room
plt.figure(figsize=(5 * plots_per_row, 4 * total_rows))

# Initial boxplot visualization
for index, column in enumerate(num_columns):
    plt.subplot(total_rows, plots_per_row, index + 1)
    sns.boxplot(y=marketing_data[column])
    plt.title(column)

# Adjust spacing between plots to prevent overlap
plt.subplots_adjust(hspace=0.5, wspace=0.3)

# Show the plot before outlier treatment
plt.show()
```

```python
# Detect and count outliers using the IQR method
for feature in marketing_data.select_dtypes(include=np.number).columns:
    q1 = marketing_data[feature].quantile(0.25)
    q3 = marketing_data[feature].quantile(0.75)
    iqr = q3 - q1
    lower_limit = q1 - 1.5 * iqr
    upper_limit = q3 + 1.5 * iqr
    outlier_rows = marketing_data[(marketing_data[feature] < lower_limit) | (marketing_data[feature] > upper_limit)]
    print(f"{feature}: {outlier_rows.shape[0]} outliers detected")
```

```python
# Cap the outliers to upper and lower thresholds
numeric_features = marketing_data.select_dtypes(include='number').columns
for attribute in numeric_features:
    q1 = marketing_data[attribute].quantile(0.25)
    q3 = marketing_data[attribute].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    marketing_data[attribute] = marketing_data[attribute].clip(lower=lower_bound, upper=upper_bound)
```

```python
# Boxplot visualization after outlier treatment
plt.figure(figsize=(5 * plots_per_row, 4 * total_rows))
for index, column in enumerate(num_columns):
    plt.subplot(total_rows, plots_per_row, index + 1)
    sns.boxplot(y=marketing_data[column])
    plt.title(column)

# Adjust spacing between plots again
plt.subplots_adjust(hspace=0.5, wspace=0.3)

# Show the plot
plt.show()
```

# 2. Missing Value detection and filling the columns

```python
# Calculates and prints the number of missing values for each column in the marketing_data DataFrame.
missing_data = marketing_data.isna().sum()
print(missing_data)
```

```python
# Checking for missing values and removing columns with >50% missing data
for feature_name in marketing_data.columns:
    total_missing = marketing_data[feature_name].isna().sum()
    missing_ratio = (total_missing / len(marketing_data)) * 100

    if total_missing > 0:
        print(f"Column '{feature_name}' contains {total_missing} missing entries.")

        if missing_ratio > 50:
            print(f"Missing data in '{feature_name}' is {missing_ratio:.2f}%. Removing this column.")
            marketing_data.drop(feature_name, axis=1, inplace=True)
```

```python
# Fills missing values in the 'Income' column of marketing_data with the mean value of that column.
marketing_data['Income'] = marketing_data['Income'].fillna(marketing_data['Income'].mean())
```

```python
# Calculates and prints the total number of missing (null) values in each column of the marketing_data DataFrame after filling the income column.
missing_value = marketing_data.isnull().sum()
print(missing_value)
```

# 3. Standardizing Unusual Entries in 'Marital_Status' Column

```python
# Standardizing unusual entries in the 'Marital_Status' column by grouping them under 'Alone'
print(marketing_data['Marital_Status'].value_counts())

marketing_data['Marital_Status'] = marketing_data['Marital_Status'].replace({
    'YOLO': 'Alone',
    'Absurd': 'Alone',
    'Alone': 'Alone'
})

print(marketing_data['Marital_Status'].value_counts())
```

#4. Converting 'Dt_Customer' Column to Datetime Format for Time-Based Analysis

```python
# Converting the 'Dt_Customer' column to datetime format for further time-based analysis
print(marketing_data['Dt_Customer'].value_counts())

marketing_data['Dt_Customer'] = pd.to_datetime(
    marketing_data['Dt_Customer'], format='%d-%m-%Y', errors='coerce'
)

print(marketing_data['Dt_Customer'])
```

#5. Extracting Year from 'Dt_Customer' and Storing in 'Customer_Join_Year' Column

```python
# Extracting the year from 'Dt_Customer' and storing it in a new column 'Customer_Join_Year'
marketing_data['Customer_Join_Year'] = marketing_data['Dt_Customer'].dt.year

print(marketing_data['Customer_Join_Year'])
```

#6. Extracting Month from 'Dt_Customer' and Storing in 'Customer_Join_Month' Column

```python
# Extracting the month from 'Dt_Customer' and creating a new column 'Customer_Join_Month'
marketing_data['Customer_Join_Month'] = marketing_data['Dt_Customer'].dt.month

print(marketing_data['Customer_Join_Month'])
```

#7. Creating 'Customer_Join_Period' Column Representing Join Date by Month and Year

```python
# Creating a new column that represents the customer's join date by month and year (e.g., '2023-05')
marketing_data['Customer_Join_Period'] = marketing_data['Dt_Customer'].dt.to_period('M')

print(marketing_data['Customer_Join_Period'])
```

#8. Extracting Day of the Month from 'Dt_Customer' and Storing in 'Join_Day' Column

```python
# Extracting the day of the month from the 'Dt_Customer' date and storing it in a new column
marketing_data['Join_Day'] = marketing_data['Dt_Customer'].dt.day

print(marketing_data['Join_Day'])
```

# 9. Dropping 'Complain' Column After Reviewing Its Value Counts


```python
marketing_data['Complain'].value_counts()
marketing_data.drop('Complain', axis=1, inplace=True)
```
